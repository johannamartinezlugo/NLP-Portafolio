```python
from flair.models import TextClassifier
from flair.data import Sentence

classifier = TextClassifier.load('en-sentiment')
sentence = Sentence('tweets_Paro.xlsx')

print('tweets_Paro.xlsx')

classifier.predict(sentence, multi_class_prob=False)
l = sentence.labels

l[0].value, l[0].score
```

```python
classifier = TextClassifier.load('sentiment')
sentence = Sentence("El paro no PARA")
classifier.predict(sentence)
sentence.labels
```

```python
classifier = TextClassifier.load('sentiment')

def sent_flair(texto):
    sentence = Sentence(texto)
    classifier.predict(sentence)
    result = sentence.labels[0]
    label = result.value
    score = result.score
    if label == 'POSITIVE':
        return score
    if label == 'NEGATIVE':
        return -1 * score
    return score

data['sent_flair'] = data['text'].apply(lambda valor: sent_flair(valor))
data.head()
```

```python
data[['text', 'sent_flair']].sort_values('sent_flair')
```
